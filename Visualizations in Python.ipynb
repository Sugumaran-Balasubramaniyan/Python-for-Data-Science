# <font color='blus'> Plot lines : `plt.plot`

```python
from matplotlib import pyplot as plt
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp   = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
# draw a curve; years on the x-axis, GDP (GDP) on the y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
# ajoute un titre
plt.title("GDP values")
# ajoute une étiquette sur l'axe des y
plt.ylabel("Billions of dollars")
plt.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
# Multiple calls to plt.plot
# leads to represents multiple series on the same graph.

# Solid green line :
plt.plot(xs, variance, 'g-', label='variance')

# Dashed blue line :
plt.plot(xs, total_error, 'b:', label='total error')

# Dashed red line with another style :
plt.plot(xs, bias_squared, 'r-.', label='bias^2')

# Because we have assigned labels to each series,
# we have an automatic caption:
# loc=9 means "top center".
plt.legend(loc=9)
plt.xlabel("Model complexity")# Do not confuse with algorithmic complexity 
plt.xticks([])
plt.ylabel("Errors")
plt.title("Tradeoff bias - Variance")
plt.show()
```

# <font color='blus'> Bar charts : `plt.bar`

> Bar charts are used to compare values.

```python
movies     = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
# Dram bars with :
# X : movies titles
# Y  : nb. rewards
plt.bar(range(len(movies)), num_oscars)
plt.title("Rewarded moovies") # ajoute un titre
plt.ylabel("Nb. Oscars") # étiquette de l'axe y
# étiquette de l'axe x avec les noms des films au centre des barres
plt.xticks(range(len(movies)), movies)
plt.show()
```

# <font color='blus'> Histograms : `plt.hist`

> Use to draw distributions.
>
> This function creates groups (categories) of values and plot the number of elements in each of the classes.
>
> Groups are created based on the values ranges. Groups are also named `bins`.

```python
import numpy as np

x = np.random.randn(1000)
print(x.mean(), x.std(), x.min(), x.max())
```

```python
#plt.legend("Legend")
plt.title("Filled Histogram ")
plt.hist(x, 25, range=(-4., 4.), histtype='stepfilled', align='mid', color='c', label='Normal distribution', density=True);
```

```python
#plt.legend("Legend")
plt.title("Filled Histogram ")
plt.hist(x, 25, range=(-4., 4.), align='mid', color='c', label='Normal distribution', density=False);
```

```python
from collections import Counter
list_score = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
# Socres classification per déciles, 100 is assigned class [90,100]
histogram = Counter(min(score // 10 * 10, 90) for score in list_score)
plt.bar([x + 5 for x in histogram.keys()], # Bars are shifted on the right side
    histogram.values(), # Assign each bar a value
    10, # width of 10 units
    edgecolor=(0, 0, 0)# blach lines for bars
       ) ;# Avoiding return value
```

# <font color='blus'> Cloud of points

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)
# étiquette chaque point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
    xy=(friend_count, minute_count),# Locate label close to the point
    xytext=(5, -5),# not to close, with a light shift
    textcoords='offset points')
plt.title("Minutes per day / Nb. friends")
plt.xlabel("Nb. friends")
plt.ylabel("Dayly time on site (min)")
plt.show()                 
```
